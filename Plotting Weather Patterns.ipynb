```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd


def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(25,'9b8a3493628de4a545ca348012ca3563e0a85b955446eef6f8fdafb5')
```

```python
import numpy as np
dfx = pd.read_csv('data/C2A2_data/BinnedCsvs_d25/9b8a3493628de4a545ca348012ca3563e0a85b955446eef6f8fdafb5.csv')

date = pd.DatetimeIndex(dfx['Date'])

dfx['Month'] = date.month
dfx['Day'] = date.day
dfx = dfx[~((dfx['Month'] == 2 ) & (dfx['Day'] == 29))]
dfx = dfx.drop('ID',1).sort(['Date','Element']).reset_index(drop = True)

df2015 = dfx[list(dfx[dfx['Date']=='2015-01-01'].index)[0]:]
df = dfx[0:list(dfx[dfx['Date']=='2015-01-01'].index)[0]]


_min = df.groupby(['Month','Day'])['Data_Value'].min()/10
_max = df.groupby(['Month','Day'])['Data_Value'].max()/10
min2015 = df2015.groupby(['Month','Day'])['Data_Value'].min()/10
max2015 = df2015.groupby(['Month','Day'])['Data_Value'].max()/10

#2015 record broken
broken_min = np.where(min2015.to_frame()['Data_Value']<_min.to_frame()['Data_Value'])
broken_max = np.where(max2015.to_frame()['Data_Value']>_max.to_frame()['Data_Value'])

broken_max
```

```python
linear_min = np.array(_min)
linear_max = np.array(_max)

plt.figure()
plt.plot(linear_min, '-', label ='Record Low')
plt.plot(linear_max, '-', label = 'Record High')
plt.fill_between(range(len(linear_max)), linear_min, linear_max, facecolor='grey', alpha=0.15)

plt.scatter(broken_min,min2015.iloc[broken_min],s = 10, color = 'darkblue', label = 'Broken Low')
plt.scatter(broken_max,max2015.iloc[broken_max],s = 10, color = 'darkred', label ='Broken High')

plt.legend(loc = 8)

xticks = (pd.date_range('31/12/2014','30/11/2015', freq = 'M')+pd.Timedelta('1D')).strftime('%-j').astype(int)
xticks_labels = (pd.date_range('31/12/2014','30/11/2015', freq = 'M')+pd.Timedelta('1D')).strftime('%b')

ax = plt.gca()
ax.set_xlim(1,365)
ax.set_xticks(xticks)
ax.set_xticklabels(xticks_labels)

for label in ax.xaxis.get_ticklabels():
    label.set_horizontalalignment('left')

y2ticks_labels = [' ', '14Â°F', '32Â°F', '50Â°F', '68Â°F', '86Â°F', '104Â°F', '122Â°F']
yticks_labels = [' ','-10Â°C', '0Â°C', '10Â°C', '20Â°C', '30Â°C', '40Â°C', '50Â°C'] 
ax.set_ylim(-20,50)
ax2=ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_yticks(ax.get_yticks()*1.8+32)
ax2.set_ylim(ax.get_yticks()[0]*1.8+32,ax.get_yticks()[7]*1.8+32)
ax2.set_yticklabels(y2ticks_labels)
ax.set_yticklabels(yticks_labels)



plt.title('Historical High & Low Temerature over 2005-2014 \nand 2015 Record-breaking Points',fontsize=10)

```
